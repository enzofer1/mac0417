```python
import matplotlib.pyplot as plt
import os
import numpy as np


files = os.listdir()
jpg_files = [f for f in files if f.endswith('.jpg')]

cgfah_files = [f for f in jpg_files if f.startswith('cgfah')]
exsbf_files = [f for f in jpg_files if f.startswith('exsbf')]

def plot_images(image_files, title):
    num_images = len(image_files)
    rows = int(np.ceil(np.sqrt(num_images)))
    cols = int(np.ceil(num_images / rows))
    fig, axes = plt.subplots(rows, cols, figsize=(12, 12))

    for i, file in enumerate(image_files):
        image = plt.imread(file)
        ax = axes[i // cols, i % cols]
        ax.imshow(image)
        ax.axis('off')
    for i in range(num_images, rows * cols):
        axes[i // cols, i % cols].axis('off')

    fig.suptitle(title)


    plt.tight_layout()

    plt.show()

plot_images(cgfah_files, 'Colher, garfo, fio dental, abridor e hashi')

```

```python
plot_images(exsbf_files, 'Esponja, xcara, sabonete, bomba de mate e faca')
```

```python

class_names = ['colher', 'garfo', 'fio dental', 'abridor', 'hashi', 'esponja', 'caneca', 'sabonete', 'bomba para mate', 'faca']

def calculate_size(image_files):
    total_size = sum(os.path.getsize(file) for file in image_files)
    return total_size / (1024 * 1024)

num_images = len(jpg_files)

total_size_mb = calculate_size(jpg_files)

if jpg_files:
    first_image = plt.imread(jpg_files[0])
    resolution = first_image.shape[:2]
else:
    resolution = (0, 0)
print("=" * 34)
print("Tabela Global Sumria:")
print(" Nome das classes:", ", ".join(class_names))
print(" Number of classes:", len(class_names))
print(" Number of images:", num_images)
print(" Size (bytes):", f"{total_size_mb:.2f} MB")
print(" Resolution:", resolution[0], "rows and", resolution[1], "columns")
print("=" * 34)
```
